```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Dataframe').getOrCreate()
```

### Read the dataset

```python
# Method 1
df_pyspark = spark.read.option('header', 'true').csv('Book1.csv')
df_pyspark.show()
```

```python
# Check the schema
df_pyspark.printSchema()
```

```python
# Use inferSchema=True
df_pyspark = spark.read.option('header', 'true').csv('Book1.csv', inferSchema=True)
df_pyspark.printSchema()
```

```python
# Method 2
df_pyspark = spark.read.csv('Book1.csv', header=True, inferSchema=True)
df_pyspark.show()
```

### Selecting Columns And Indexing

```python
df_pyspark.columns
```

```python
df_pyspark.head(3)
```

```python
# Selecting a column
df_pyspark.select('Name').show()
```

```python
# Selecting multiple columns
df_pyspark.select(['Name', 'Experience'])
```

```python
# Check data types
df_pyspark.dtypes
```

```python
# Describe option similar to Pandas
df_pyspark.describe().show()
```

### Add Column

```python
df_pyspark = df_pyspark.withColumn('Experience after 2 yrs', df_pyspark['Experience']+2)
```

```python
df_pyspark.show()
```

### Drop column

```python
df_pyspark = df_pyspark.drop('Experience after 2 yrs')
```

```python
df_pyspark.show()
```

### Rename the column

```python
df_pyspark.withColumnRenamed('Experience','Exp').show()
```
